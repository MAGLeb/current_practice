```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_prob(number_of_people):
    numerator = np.math.factorial(365)
    denominator = np.math.factorial(365 - number_of_people) * (365 ** number_of_people)
    return 1 - (numerator / denominator)
```

```python
probability = []
numbers_of_people = []

for number in range(100):
    prob = find_prob(number)
    probability.append(prob)
    numbers_of_people.append(number)
    
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
fig.suptitle("Probability of two people have the same day of birthday", fontsize=32)
ax.set_xlabel("Number of people", fontsize=16)
ax.set_ylabel("Probability", fontsize=16)
ax.plot(numbers_of_people, probability, linewidth=4)
ax.plot(23, probability[23], 'ro', label='23 people / 0.5072', markersize=12)
ax.plot(30, probability[30], 'go', label='30 people / 0.7063', markersize=12)
ax.plot(50, probability[50], 'yo', label='50 people / 0.9703', markersize=12)
plt.legend(fontsize=14)
```

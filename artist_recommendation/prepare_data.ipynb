```python
%%time

albums_ids = []

w = open('./entities/albums.json', 'w')
with open('./entities/albums.idomaar', 'r') as f:
    for line in f:
        line = line.split('\t')
            
        if line[0]:

            albums_ids.append(line[1])

            w.write(line[3] + '\n')
   

w.seek(0, 0)
w.close()
```

```python
def removeQuoteMarks(example):
    counter = 0

    for letter in example:
        if letter == '"':
            counter += 1
    
    if '/' in example:
        example = example.replace('/', '|')
        
    if '\\' in example:
        example = example.replace('\\', '|')
            
    if counter > 8:
        j = 0
        array = list(example)
        for i in range(len(array)):
            if array[i] == '"':
                j += 1
            if (j > 7 and j != counter) and array[i] == '"':
                array[i] = "\'"
                

        example = ''.join(array)
        
    return example
```

С данными albums столкнулся с проблемой. Есть некоторые объекты, в которых используются ковычки в названии альбома, что не дает верно считать JSON. Написал маленький парсер, который заменял лишние двойные кавычки на одинарные.

```python
%%time

w = open('./entities/albums_removed.json', 'w')

with open('./entities/albums.json', 'r') as f:
    for line in f:
        w.write(removeQuoteMarks(line))
            
w.seek(0, 0)
w.close()
        

        
        

```

```python
%%time

sessions_ids = []
num_stracks = []
playtime = []
user_ids = []

w = open('./relations/sessions.json', 'w')
with open('./relations/sessions.idomaar', 'r') as f:
    for line in f:
        line = line.split('\t')
        
        sessions_ids.append(int(line[1]))
        
        body = line[3].split(' ')
        num_stracks.append(json.loads(body[0])['numtracks'])
        playtime.append(json.loads(body[0])['playtime'])
        user_ids.append(json.loads(body[1])['subjects'][0]['id'])
        
        interm = json.loads(body[1])['objects']
        
        for item in interm:
            item['session_id'] = int(line[1])
            w.write(json.dumps(item) + '\n')
   

w.seek(0, 0)
w.close()
```
